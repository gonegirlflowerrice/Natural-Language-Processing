```python
import spacy
nlp_sm = spacy.load("en_core_web_sm")
```

```python
#######################RATING STAR 1=================================================
```

```python
# Task: Randomly select 50 reviews (one from each business) of rating 1

# Filter 1 star reviews from data
oneStarReviews = df[df['stars']==1]

# Display 1 star reviews
display(oneStarReviews)
```

```python
# Group 1 star ratings by business ID
groupedByBusinessID = oneStarReviews.groupby('business_id').sample(n=1)

len(groupedByBusinessID)
```

```python
# Select random 50 samples
randomReview = groupedByBusinessID.sample(n=50)

# Display Businesses
display(randomReview)
```

```python
pd.options.display.max_colwidth = 500
print(randomReview.text)
```

```python
# Display length (50 reviews)
display(len(randomReview))
```

```python
# Display 50 reviews unique id
display(randomReview["business_id"].unique())
display(len(randomReview["business_id"].unique()))
```

```python
# Task: Extract the top-10 most frequent noun-adjective pairs from the sentences in these selected reviews

noun_adj_pair_list = []
for i in randomReview["text"]:
    doc = nlp_sm(i)
    for i,token in enumerate(doc):
        if token.pos_ not in ('NOUN','PROPN'):
            continue
        for j in range(i+1,len(doc)):
            if doc[j].pos_ == 'ADJ':
                noun_adj_pair_list.append((str(token),str(doc[j])))
                break
noun_adj_pair_list
```

```python
from collections import Counter
counter_pairs = Counter(noun_adj_pair_list)
print(counter_pairs.most_common(10))
```
